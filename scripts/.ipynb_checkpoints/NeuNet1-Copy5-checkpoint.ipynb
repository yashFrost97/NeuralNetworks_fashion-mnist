```python
# Read Fashion MNIST dataset

import util_mnist_reader
X_train, y_train = util_mnist_reader.load_mnist('../data/fashion', kind='train')
X_test, y_test = util_mnist_reader.load_mnist('../data/fashion', kind='t10k')

# Your code goes here . . .

X_train = X_train / 255.0
X_test = X_test / 255.0

print(X_train.shape)
print(y_train.shape)

```

```python
encoded = np.zeros((60000, 10))

for i in range(60000):
    encoded[i, y_train[i]] = 1
    
```

```python
print(encoded.shape)
print(encoded)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

# X_train = np.reshape(X_train, [60000, 28, 28])

# initialize the values
hidden_nodes = 150 # no of hidden nodes
weight_l1 = np.random.rand(784, hidden_nodes)
bias_l1 = np.random.rand(hidden_nodes,)

# print("Initial Weights layer 1")
# print(weight_l1)
weight_l2 = np.random.rand(hidden_nodes, 10)
bias_l2 = np.random.rand(10,)

# print("Initial Weights layer 2")
# print(weight_l2)


losstrack = []
training_errors = []
test_wrong = []

l_rate = 0.01
epochs = 100

print(weight_l1.shape)
print(bias_l1.shape)

print(weight_l2.shape)
print(bias_l2.shape)

def sigmoid(z):
    ans = 1/(1 + np.exp(-z.astype(float)))
    return ans

def softmax(z):
#     e_x = np.exp(z - np.max(z))
#     return e_x / np.sum(e_x)
#     z = z - z.max(axis = None, keepdims = True)
#     y = np.exp(z)
#     return y / y.sum(axis = None, keepdims = True)
    temp = np.exp(z - np.max(z, axis = 0, keepdims = True))
    return temp / np.sum(temp, axis = 0, keepdims = True)

def s_der(x):
    return x*(1-x)
```

```python
def loss_fn(train, final):
    loss = np.sum(-np.multiply(final, np.log(train + 1e-20)))
    loss = loss / 60000
    return loss
```

```python
# import sys
# np.set_printoptions(threshold = sys.maxsize)


for epoch in range(epochs):
    z_1 = (np.dot(weight_l1.T, X_train.T).T + bias_l1).T
    a_1 = sigmoid(z_1)
    z_2 = (np.dot(weight_l2.T, a_1).T + bias_l2).T
    a_2 = softmax(z_2)
    print(a_2.shape)
    loss = loss_fn(a_2, encoded.T)
    losstrack.append(loss)
    print(f"loss = {loss}")
    
#   back propagation starts here
    error_y = (a_2.T - encoded) / 60000.0
    sigma_deri = s_der(a_1.T)
    
#   DONT TOUCH THIS!!
    part1 = np.dot(error_y, weight_l2.T)
    t1 = np.multiply(part1, sigma_deri)
    dw_1 = np.dot(t1.T, X_train)
    dw_1 = dw_1.T
    #print(dw_1.shape)
    dw_2 = np.dot(error_y.T, a_1.T).T
    #print(dw_2.shape)

    
    #t1 = np.dot(((error_y.T) / 60000), weight_l2.T).T
   # t2 = np.sum(np.multiply(t1, sigma_deri), axis = 1)
    dbias_1 = np.sum(t1,axis=0)
    
    dbias_2 = np.sum((error_y), axis = 0)
#     weight_l1 = np.subtract(weight_l1, np.multiply(l_rate, dw_1))
#     weight_l2 = np.subtract(weight_l2, np.multiply(l_rate, dw_2))
    weight_l1 -= l_rate * dw_1
    weight_l2 -= l_rate * dw_2
    bias_l1 = bias_l1 - l_rate * dbias_1
    bias_l2 = bias_l2 - l_rate * dbias_2
    
    prediction = []
    r = a_2.T
    for i in r:
        prediction.append(np.argmax(i))
    accuracy = (y_train==prediction).mean()
    
    print(f"epoch = {epoch}")
    print(dbias_2.shape)
    print(accuracy)

plt.plot(losstrack)
```
