```python
# Read Fashion MNIST dataset

import util_mnist_reader
X_train, y_train = util_mnist_reader.load_mnist('../data/fashion', kind='train')
X_test, y_test = util_mnist_reader.load_mnist('../data/fashion', kind='t10k')

# Your code goes here . . .

print(X_train)

X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

print(X_train.shape)

# X_train = np.reshape(X_train, [60000, 28, 28])

print(X_train.shape)

# initialize the values
hidden_nodes = 100 # no of hidden nodes
weight_l1 = np.random.rand(784, hidden_nodes)
bias_l1 = np.random.rand(hidden_nodes, 1)

weight_l2 = np.random.rand(hidden_nodes, 10)
bias_l2 = np.random.rand(10, 1)

losstrack = []
training_errors = []
test_wrong = []

l_rate = 0.01
epochs = 300

print(weight_l1.shape)
print(bias_l1.shape)

print(weight_l2.shape)
print(bias_l2.shape)

def sigmoid(z):
    ans = 1/(1 + np.exp(-z))
    return ans

def softmax(z):
    e_x = np.exp(z - np.max(z))
    return e_x / np.sum(e_x)
#     z = z - z.max(axis = None, keepdims = True)
#     y = np.exp(z)
#     return y / y.sum(axis = None, keepdims = True)


# def feed_forward():
#     z_1 = np.dot(weights_l1.T, X_train) + bias_l1
#     a_1 = sigmoid(z_1)
#     z_2 = np.dot(weights_l2.T, a_1) + bias_l2
#     a_2 = softmax(z_2)
#     # return a_2

# def sigmoid_prime(a):
#     return np.dot(a, (1 - a).T)
    
# def feedbackward():
#     dw_2 = np.dot(a_2 - y_train, a_1)
#     w_2 = w_2 + dw_2
#     a = sigmoid_prime(a_1)
#     a = np.dot(a, X_train)
#     dw_1 = np.dot(dw_2, a)
#     dw_1 = w_1 + dw_1
    
```

```python
def cross_entropy(train, final):
    loss = np.sum(-np.multiply(final, np.log(train + 1e-20)))
    return loss
```

```python
for epoch in range(epochs):
    z_1 = np.dot(weight_l1.T, X_train.T)
    a_1 = sigmoid(z_1)
    z_2 = np.dot(weight_l2.T, a_1)
    a_2 = softmax(z_2)
    
    loss = cross_entropy(a_2, y_train)
    losstrack.append(loss)
    print(f"loss = {loss}")
    
#     part_1 = np.dot((a_2 - y_train).T, weight_l2.T)
    
#     sig_der = np.multiply(a_1, (1-a_1).T)
#     print(sig_der.shape)
#     part_2 = np.multiply(sig_der, X_train)
#     dw_1 = np.multiply(part_1, part_2)
    dw_1 = np.dot(np.subtract(a_2, y_train.T).T/60000.0, weight_l2.T)
    dw_1 = np.dot(dw_1.T, a_1.T)
    dw_1 = np.dot(dw_1, (1 - a_1))
    dw_1 = np.dot(dw_1, X_train)
    
    dw_2 = np.dot(np.subtract(a_2, y_train.T)/60000.0, a_1.T)
    
    weight_l1 = weight_l1 - l_rate * dw_1.T
    weight_l2 = weight_l2 - l_rate * dw_2.T
    
    print(f"epoch = {epoch}")
plt.plot(losstrack)

# Training accuracy
train_output = []
for i in range(len(y_test)):
    max = 0
    k = 0
    for j in range(10):
        if(a_2[j][i] > max):
            max = a_2[j][i]
            k = j
    train_output.append(k)
    
train_output = np.array(train_output)
print(train_output.shape)

equals = 0
for i in range(len(y_train)):
    if(y_train[i] == int(train_output[i])):
        equals += 1
        
acc = equals/len(y_train)
print(train_output)
print(f"accuracy = {acc}")        
```
